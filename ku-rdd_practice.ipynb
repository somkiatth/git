```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Practice1').getOrCreate()
```

```python
sc = spark.sparkContext
```

```python
score_rdd = sc.textFile('/content/drive/MyDrive/Datasets/rdd_problem.csv')
```

```python
score_split_rdd = score_rdd.map(lambda x: [x.split(',')[0],
                                           x.split(',')[1],
                                           int(x.split(',')[2])
]).cache()
score_split_rdd.take(5)
```

# 1

```python
score_split_rdd.filter(lambda x: x[2] == 100).count()
```

# 2

```python
score_split_rdd.map(lambda x: (x[1],x[2]))\
               .groupByKey()\
               .map(lambda x: (x[0], len(x[1])) ).take(5)
```

# 3

```python
score_split_rdd.map(lambda x: (x[1],(x[2], 1)))\
               .reduceByKey(lambda x,y: (x[0]+y[0], x[1]+y[1]))\
               .map(lambda x: (x[0], x[1][0]/x[1][1]))\
               .collect()
```

```python
score_split_rdd.map(lambda x: (x[1],x[2]))\
               .groupByKey()\
               .mapValues(list)\
               .collect()
```
